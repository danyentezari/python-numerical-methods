```python

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
# Note: the sympy symbols will prevent default variable behavior
from sympy import *
x, y, z, t = symbols('x y z t')
```

```python
def eulerMethod(f, x_0, y_0, h=0.25, N=10):
    """
    :param lambda f: unknown (multivariate) function
    :param int x_0: initial value (x)
    :param int y_0: initial value (y)
    :param int h: step size (i.e; x_n - x_0)
    :param int N: number of steps
    """

    x_1 = x_0 + h
    y_1 = y_0 + ((1/2) * h) * f(x_0, y_0)

    k_1 = h * f(x_1, y_1)
    y_2 = y_0 + (1/2) * k_1
    x_2 = x_1 + h


    return x_2, y_2


f = lambda x, y: x*2
x_0 = 1
y_0 = 1
x_1, y_1 = eulerMethod(f, x_0, y_0)
x_2, y_2 = eulerMethod(f, x_1, y_1)


plt.scatter(x_0, y_0, color='b', s=50)
plt.scatter(x_1, y_1, color='b', s=50)
plt.scatter(x_2, y_2, color='b', s=50)
plt.show()

```

```python
def rungeKutta(f, x_0, y_0, h, N=10):
    """
    :param lambda f: uknown function
    :param int x_0: initial value (x)
    :param int y_0: initial value (y)
    :param int h: step size (i.e; x_n - x_0)
    :param int N: number of steps
    """

    x = [
        x_0,
        x_0 + h,
        x_0 + 3 * h,
        x_0 + 4 * h,
        x_0 + 5 * h,
        x_0 + 6 * h,
        x_0 + 7 * h,
        x_0 + 8 * h,
        x_0 + 9 * h,
        x_0 + 10 * h
    ]

    for i in range(1, N):
        k1 = h * f(x[i], y[i])

    return x[0], k1
```

```python
"""
Example of plotting with matplotlib
Reference: [1, https://github.com/CambridgeUniversityPress/NumericalMethodsPhysicsWithPython/blob/master/codes/plotex.py]
"""

def plotex(cxs,cys,dxs,dys):
    plt.xlabel('x', fontsize=20)
    plt.ylabel('f(x)', fontsize=20)
    plt.plot(cxs, cys, 'r-', label='quadratic')
    plt.plot(dxs, dys, 'b--^', label='other function')
    plt.legend()
    plt.show()

cxs = [0.1*i for i in range(60)]
cys = [x**2 for x in cxs]
dxs = [i for i in range(7)]
dys = [x**1.8 - 0.5 for x in dxs]

plotex(cxs, cys, dxs, dys)
```

```python
def slope(x): 
    return 2*x

def line(x, x1, y1):
    return slope(x1)*(x - x1) + y1

plt.xlabel('x', fontsize=20)
plt.ylabel('f(x)', fontsize=20)

y = lambda x: x**2
gradient = (40, y(40))
gradient = {
    'x': gradient[0],
    'y': gradient[1]
}

domain = np.linspace(0,70,100)
xrange = np.linspace(gradient['x']-10, gradient['x']+10, 10)



plt.plot(domain, y(domain), 'r-', label='quadratic')
plt.scatter(gradient['x'], gradient['y'], color='b', s=50)
plt.plot(xrange, line(xrange, gradient['x'], gradient['y']), 'b--', linewidth = 2)
plt.show()
```
